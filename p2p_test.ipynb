```python
import numpy as np
import pandas as pd
```

```python

test_df = pd.read_csv("test.csv")
```

```python
test_df
```

```python
dates = ['origination_date','last_payment_date','next_payment_due_date']
for col in dates:
    test_df[col] = pd.to_datetime(test_df[col])
```

```python
test_df['org_date_month'] = test_df['origination_date'].dt.month
```

```python
test_df['org_date_year'] = test_df['origination_date'].dt.year
```

```python
test_df['lst_payment_month'] = test_df['last_payment_date'].dt.month
test_df['lst_payment_year'] = test_df['last_payment_date'].dt.year
```

```python
test_df['pay_period']=test_df['last_payment_date']-test_df['origination_date']
```

```python
test_df = test_df.drop(['origination_date','last_payment_date','next_payment_due_date'],axis=1)
```

```python
test_df
```

```python
cols = list(test_df.columns)
```

```python
re_cols = ['loan_number',
 'amount_borrowed',
 'term',
 'borrower_rate',
 'installment',
 'grade',
 'org_date_month',
 'org_date_year',
 'lst_payment_month',
 'lst_payment_year',
 'pay_period',
 'listing_title',
 'principal_balance',
 'principal_paid',
 'interest_paid',
 'late_fees_paid',
 'days_past_due',
 'data_source']
```

```python
test_df = test_df[re_cols]
```

```python
test_df
```

```python
X = test_df.drop(['loan_number','data_source','org_date_month','lst_payment_month'],axis=1)
X
```

```python
X['pay_period'] = X['pay_period'].astype('timedelta64[D]').astype(int)
X
```

```python
object_cols = ['grade', 'listing_title']
```

```python
from sklearn.preprocessing import OneHotEncoder

# Apply one-hot encoder to each column with categorical data
OH_encoder = OneHotEncoder(handle_unknown='ignore', sparse=False)
OH_cols_train = pd.DataFrame(OH_encoder.fit_transform(X[object_cols]))


# One-hot encoding removed index; put it back
OH_cols_train.index = X.index


# Remove categorical columns (will replace with one-hot encoding)
X = X.drop(object_cols, axis=1)


# Add one-hot encoded columns to numerical features
X = pd.concat([X, OH_cols_train], axis=1)
```

```python
X.to_csv('test_modified.csv',index=False)
```
